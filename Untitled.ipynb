```python
#### TESTING
import math 
import random
```

```python
def printMatrix(mat):
    print("MATRIX:")
    for i in range(9):
        for j in range(9):
            print(mat[i][j], end=" ")
        print()
        
        
def getRow(i, mat):
    return mat[i]

def getColumn(i, mat):
    return [x[i] for x in mat]

def getSubSquareValues(i,j, mat):
    startingX , startingY = 0,0
    matLen = int(math.sqrt(len(mat)))
    
    startingX = (i // matLen) * matLen  # integer part of i divided by matLen
    startingY = (j // matLen) * matLen  # integer part of j divided by matLen
   
    empArr = []
    for i in range(matLen):
        for j in range(matLen):
            empArr.append(mat[startingX+i][startingY+j])
    
    return empArr
```

```python
#random row constraint
rowConstrainArray  = [9, 3, 2, 4, 7, 1, 6, 5, 8]

#generate empty matrix 
matrix = [[0]*9]*9
matrix[0] = rowConstrainArray
```

```python
#### print the empty matrix with constraints
printMatrix(matrix)
```

```python
#### test functions getRow getColumn
print("1st row")
print(getRow(0, matrix))
print("2nd row")
print(getRow(1, matrix))

print("1st column")
print(getColumn(0, matrix))
print("2nd column")
print(getColumn(1, matrix))

###generate random matrix for testing
tmpMat = [[random.randint(1, 9) for t in range(9)] for x in range(9)]
printMatrix(tmpMat)

###test the functions
print("subsquare Testing")
for i in range(9):
    for j in range(9):
        print("Subsquare values for ", i , j , " are: ", getSubSquareValues(i,j, tmpMat))
```

```python
#### generation

done = False

while not done:

    candidates = []
    
    #construct potential candidates
    for i in range(len(matrix)):
        candRow = []
        for j in range(len(matrix)):
            if matrix[i][j] == 0:
                candRow.append(list(range(1,10)))
            else:
                candRow.append([0]*9)
        candidates.append(candRow)
    
    #### print candidates generation
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            print(i , j , candidates[i][j])
    
    #### print candidates generation
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] == 0:
                rowValues = getRow(i, matrix)
                colValues = getColumn(j, matrix)
                subSquare = getSubSquareValues(i,j, matrix)

                for x in set(rowValues + colValues + subSquare):
                    if x in candidates[i][j]:
                        candidates[i][j].remove(x)
                    
    #### print candidates after removing
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            print(i , j , candidates[i][j])
 

    ###generate a list of (i,j) where the value is 0 for picking a random candidate
    tmpList = []
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            if matrix[i][j] == 0:
                tmpList.append((i,j))

    #print the list
    for i in range(len(tmpList)):
        print(tmpList[i])
    
    randPos = 0
    while True:
        print(len(tmpList))
        randPos = random.randint(0, len(tmpList)-1)
        if matrix[tmpList[randPos][0]][tmpList[randPos][0]] == 0:
            break
    
    print(randPos)
    print(tmpList[randPos])
    #### pick a rand candidate for that position
    randCandidatePos = random.randint(0, len(candidates[tmpList[randPos][0]][tmpList[randPos][1]])-1)
    print(candidates[tmpList[randPos][0]][tmpList[randPos][1]])
    print(candidates[tmpList[randPos][0]][tmpList[randPos][1]][randCandidatePos])
    matrix[tmpList[randPos][0]][tmpList[randPos][1]] = candidates[tmpList[randPos][0]][tmpList[randPos][1]][randCandidatePos]
    
    candidates[tmpList[randPos][0]][tmpList[randPos][1]].remove(candidates[tmpList[randPos][0]][tmpList[randPos][1]][randCandidatePos])
    
```
